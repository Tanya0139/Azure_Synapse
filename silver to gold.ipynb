## Transformation for all tables

```python
table_name = []
for i in dbutils.fs.ls('/mnt/silver/SalesLT'):
  table_name.append(i.name.split('/')[0])
```

```python
table_name
```

```python
for n in table_name:
    path = '/mnt/silver/SalesLT/'+n
    print(path)
    df = spark.read.format('delta').option("header", "true").option("inferSchema", "true").load(path)
    cn = df.columns
    for ocn in cn :
        ncn = "".join(["_" + char if char.isupper() and not ocn[i - 1].isupper() else char for i, char in enumerate(ocn)]).lstrip("_")
        df = df.withColumnRenamed(ocn, ncn)
    op = '/mnt/gold/SalesLT/'+n+'/'
    df.write.format('delta').mode('overwrite').save(op)
```
